```python
#temperature(0,17,21,25,35)、steps(0,4000,6000,15000,30000)、sunlight(0,2,5,8,14)をカテゴライズして比較してみよう
```

```python
for df in data:
    df["temperature_band"]=pd.cut(ma["temperature"],[0,17,21,25,35],labels=range(4),right=False)
ma["temperature_band"].value_counts()
```

```python
for df in data:
    df["steps_band"]=pd.cut(ma["steps"],[0,4000,6000,15000,30000],labels=range(4),right=False)
ma["steps_band"].value_counts()
```

```python
for df in data:
    df["sunlight_band"]=pd.cut(ma["sunlight"],[0,2,5,8,14],labels=range(4),right=False)
ma["sunlight_band"].value_counts()
```
